```python
import matplotlib.pyplot as plt
```

```python
# y=x2−x+2
x = list(range(0, 10))
y = [n ** 2 - n + 2 for n in x]
plt.plot(x, y)
plt.text(0, 5, '(0, 0)', fontsize=10, color='blue')
```

```python
# y=√x
x = list(range(0, 5))
y = [n ** 0.5 for n in x]
plt.plot(x, y)
```

```python
# y=x3
x = list(range(0, 5))
y = [n ** 3 for n in x]
plt.plot(x, y)
```

```python
# y=2x
x = list(range(0, 5))
y = [2 ** n for n in x]
plt.plot(x, y)
```

```python
#y=1/(x+1)
x = list(range(0, 5))
y = [1/(n + 1) for n in x]
plt.plot(x, y)
```

```python
# y=√x
x = list(range(0, 5))
y = [n ** 0.5 for n in x]
plt.subplot(2, 2, 1)
plt.plot(x, y)
plt.title('y = sqrt(x)')

# y=x3
x = list(range(0, 5))
y = [n ** 3 for n in x]
plt.subplot(2, 2, 2)
plt.plot(x, y)
plt.title('y = x cubed')

# y=2x
x = list(range(0, 5))
y = [2 ** n for n in x]
plt.subplot(2, 2, 3)
plt.plot(x, y)
plt.title('y = double x')

# y=1/(x+1)
x = list(range(0, 5))
y = [1/(n + 1) for n in x]
plt.subplot(2, 2, 4)
plt.plot(x, y)
plt.title('y=1/(x+1)')

plt.subplots_adjust(left=0.1,
                    bottom=0.1, 
                    right=0.9, 
                    top=0.9, 
                    wspace=0.4, 
                    hspace=0.4)
plt.show()
```

```python

# y=√x
x = list(range(0, 5))
y_sqrt_x = [n ** 0.5 for n in x]
plt.plot(x, y_sqrt_x, color = 'red', label = 'Sqrt X')


# y=x3
x = list(range(0, 5))
y_x_cubed = [n ** 3 for n in x]
plt.plot(x, y_x_cubed, color = 'green', label = 'Cubed X')

# y=2x
x = list(range(0, 5))
y_2x = [2 ** n for n in x]
plt.plot(x, y_2x, color = 'blue', label = 'Two X')

# y=1/(x+1)
x = list(range(0, 5))
y_last = [1/(n + 1) for n in x]
plt.plot(x, y_last, color = 'yellow', label = '1 / (x + 1)' )

plt.legend()

```
