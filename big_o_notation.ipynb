```python
import math
import matplotlib.pyplot as plt

# y=0n+1 and label the curve "O(1)"
x = list(range(1, 4))
y1 = [0 * n + 1 for n in x]
plt.plot(x, y1, color = 'red', linestyle = '--',label = 'O(1)')

# y=log(n) and label the curve "O(log n)"
x = list(range(1, 4))
y2 = [math.log(n) for n in x]
plt.plot(x, y2, color = 'green', linestyle = '--',label = 'O(log n)')

# y=n and label the curve "O(n)"
x = list(range(1, 4))
y3 = [n for n in x]
plt.plot(x, y3, color = 'blue', linestyle = '--',label = 'O(n)')

# y=n∗log(n) and label it "O(n log n)"
x = list(range(1, 4))
y4 = [n * math.log(n) for n in x]
plt.plot(x, y4, color = 'yellow', linestyle = '--',label = 'O(n log n)')


# y=n2 and label it "O(n^2)"
x = list(range(1, 4))
y5 = [n ** 2 for n in x]
plt.plot(x, y5, color = 'yellow', label = 'O(n^2)')

# y=2n and label it "O(2^n)"
x = list(range(1, 4))
y6 = [2 * n for n in x]
plt.plot(x, y6, color = 'blue', label = 'O(2^n)')

#y=n! and label it "O(n!)"
x = list(range(1, 4))
y7 = [math.factorial(n) for n in x]
plt.plot(x, y7, color = 'green', label = 'O(n!)')

#y=nn and label it "O(n^n)"
x = list(range(1, 4))
y8 = [n ** n for n in x]
plt.plot(x, y8, color = 'red', label = 'O(n^n)')

# Add a title
plt.title('Big O Notation')

# Add x and y axis labels
plt.xlabel('Elements')
plt.ylabel('Operations')

# Add grid and legend
plt.grid(True, ls = '--')
plt.legend()
```
